# Rotating the S12 and S16 components 

The horizontal components were intended to point north (MH1) and east (MH2). However, S12 and S16 were misaligned.

The true allignments are taken from Nunn 2020 data paper. 

The imput is the removed instrument response files, and the output has the same name with MH1 and MH2 replaced for MHN and MHE. 

```python
from obspy import read
import glob
import os
import obspy.signal
from obspy.core import stream
from obspy.signal import filter
from  obspy.signal.cross_correlation import correlate
from  obspy.signal.cross_correlation import xcorr_max
from scipy import stats
from obspy.io.xseed import Parser
import datetime
from  obspy.core import UTCDateTime
from obspy.signal.rotate import rotate2zne
```

```python
year = [1969,1970,1971,1972,1973,1974,1975,1976,1977]
station = ['S12','S16']
comp = ['MH1']
```

```python
for y in year:
    for sta in station:
        for co in comp:
            if sta == 'S12':
                MH1_az = 180
                MH2_az = 270
            if sta == 'S16':
                MH1_az = 334.5
                MH2_az = 64.5
            st = []
            fil='/Users/TheStuffofAlice/Dropbox/Apollo/A01_moonquakes/'+str(y)+'/'+str(sta)+'/'+str(co)+'/*'+co+'*resp*'
            fnames=glob.glob(fil)
            for fname in fnames:
                try:
                    print(fname)
                    MH2_f = fname.replace('MH1','MH2')
                    MHZ_f = fname.replace('MH1','MHZ')
                    st = read(fname)
                    st += read(MH2_f)
                    st += read(MHZ_f)
                    st_copy = st.copy()
                    print(st)
                    st_rotate = rotate2zne(st[0],0,MH1_az,st[1],0,MH2_az,st[2],90,0)
                    st[0].data = st_rotate[0]
                    st[1].data = st_rotate[1]
                    st[2].data = st_rotate[2]
                    print('rotated!')
                    st[0].write(fname.replace('.MH1.','.MHN.'),format="MSEED")
                    st[1].write(MH2_f.replace('.MH2.','.MHE.'),format="MSEED")
                    st[2].write(MHZ_f,format="MSEED")
                    print('saved!')
                except:
                    print('traces are not the same length becuase of data gaps!')
                
                
```
