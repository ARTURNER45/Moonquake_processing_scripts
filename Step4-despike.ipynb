# Despiking algorithm for the data 

This script calls the function 'despike function' which has two different versions of the median average filter. 

first a band pass fitler is applied to the data. 

The first relies on a strict threshold which is supplied, and the second cuts of short peaks that are above 5 * the running median ( a dynamic threshold) . The length of the median running average is 701 samples, or around 2 mins, which seems to work okay and remove the extreme peaks. 

First the hard filter is run, and then the dynamic threshold filter is run for each trace. 

The input to this script is the traces with the instrument response removed and the output is a file with the file ending 'despike; 

```python
from obspy import read
import glob
import obspy.signal
import obspy.core 
import obspy.signal 
from  obspy.signal.cross_correlation import correlate
from  obspy.signal.cross_correlation import xcorr_max
from scipy import stats
from obspy.io.xseed import Parser
import scipy as sp
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from despike_function import despike_step1
from despike_function import despike_step2
```

```python
year = ['1969','1970','1971','1972','1973','1974','1975','1976','1977']
station = ['S12','S14','S15','S16']
comp = ['MHN','MHE','MHZ']
```

```python
for y in year:
    for sta in station:
        for co in comp:
            st = []
            fil='/Users/TheStuffofAlice/Dropbox/Apollo/A01_moonquakes/'+str(y)+'/'+str(sta)+'/'+str(co)+'/*'+co+'*resp*'
            fnames=glob.glob(fil)
            for fname in fnames: 
                st = obspy.read(fname)
                print(st)
                tr = st[0]
                if tr.stats.npts <= 1:
                    continue
                tr.filt = tr.copy()
                df = tr.stats.delta
                tr.filt.filter('bandpass',freqmin = 0.25,freqmax = 3.31245,corners=2,zerophase=True)
                time = tr.times
                despike_tr1,fivemedian1 = despike_step1(tr.filt.data, m=100,kernel_size = 701,threshold = 30 )
                despike_tr2,fivemedian2 = despike_step2(despike_tr1, m=100,kernel_size = 701)
                st[0].data = despike_tr2
                print('despiked!')
                st[0].write(fname.split('.resp_peaked')[0]+'despike',format="MSEED")
                print('saved!')
                
```
