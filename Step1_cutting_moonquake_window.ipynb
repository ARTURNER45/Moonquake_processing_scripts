# Cutting the moonquake window from the Nakamura database

This is a script which reads in a text file with the start and end times of A01 moon events and then cuts the traces to these start and end times. 

The newly added Bulow events, and some events that run over to the next day, or run into another event do not have an end time, so the end of the dayly trace is taken. 

This reads in the raw data, and outputs a cut file with the filename appended '.cut' 

```python
from obspy import read
import glob
import os
import obspy.signal
from obspy.core import stream
from obspy.signal import filter
from  obspy.signal.cross_correlation import correlate
from  obspy.signal.cross_correlation import xcorr_max
from scipy import stats
from obspy.io.xseed import Parser
import datetime
from  obspy.core import UTCDateTime
```

```python
f = open('/Users/TheStuffofAlice/Dropbox/Apollo/NAKAMURA_DATABASE/A1_moonquakes.txt', "r")
for line in f:
    y=line.split()
    year = str(int(y[0]) + 1900)
    day = y[1].lstrip("0")
    tstart = y[2]
    tend = y[3]
    
    rootDir = '/Users/TheStuffofAlice/Dropbox/Apollo/A01_moonquakes/' + year 
    
    for dirName, subdirList, fileList in os.walk(rootDir):
        #print(dirName)
        tend = y[3]
        #print(tstart,tend)
        parts = dirName.split('/') 
        if len(parts) < 9:
            continue 
        comp= parts[8]
        station=parts[7]
        file_name = station +'.'+'XA'+'..'+comp+'.'+year+'.'+day
        print(file_name)
        try:
            st = read(dirName + '/' + file_name)
            tstart_format = str(tstart)[0]+str(tstart)[1] + ':'+ str(tstart)[2]+str(tstart)[3]
            startyear = year + " " + day + " " + tstart_format
            format_stattime = str(datetime.datetime.strptime(startyear, '%Y %j %H:%M'))
            d = datetime.timedelta(minutes = 30 )
            t1 = UTCDateTime(format_stattime) - d 
            if tend == '-1.':
                tend = st[0].stats.endtime
                t2 = UTCDateTime(tend)
                if t2 < t1 :
                    continue
            elif tend == '9999':
                continue
            elif int(tend) < int(tstart) :
                continue
            else:
                tend_format = str(tend)[0]+str(tend)[1] + ':'+ str(tend)[2]+str(tend)[3]
                endyear =  year + " " + day + " " + tend_format
                format_endtime = str(datetime.datetime.strptime(endyear, '%Y %j %H:%M'))
                t2 = UTCDateTime(format_endtime) + d 
                
            backup = st.copy()
            cut_stream = st.trim(t1, t2)
            cut_stream.write(dirName + '/' + file_name +".cut",format="MSEED") 
        except OSError:
            pass    
        
```
